```python
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x,y = make_blobs(n_samples=200,
                 n_features=2,
                 centers=3,
                 cluster_std=1.2) #값이 커지며 충돌이 생김
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
model = MeanShift(bandwidth=0.8)
pred = model.fit_predict(x)
np.unique(pred,return_counts=True)
plt.scatter(x[:,0],x[:,1],c=pred)
for center in model.cluster_centers_:
    plt.plot(center[0],center[1],'ro')
plt.show()
```

```python
model = MeanShift(bandwidth=1)
pred = model.fit_predict(x)
np.unique(pred,return_counts=True)
plt.scatter(x[:,0],x[:,1],c=pred)
for center in model.cluster_centers_:
    plt.plot(center[0],center[1],'ro')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = iris.data
label = iris.target
```

```python
pca = PCA(n_components=2)
pca.fit(data)
pca_data = pca.transform(data)
```

```python
eb = estimate_bandwidth(pca_data)
print(eb)
```

```python
model = MeanShift(bandwidth=eb)
pred = model.fit_predict(pca_data)
```

```python
np.unique(pred,return_counts=True)
```

```python
plt.scatter(pca_data[:,0],pca_data[:,1],c=pred)
plt.scatter(pca_data[:,0],pca_data[:,1],c=label, s=2.0)
plt.show()
```

```python
model = MeanShift(bandwidth=0.77)
pred = model.fit_predict(pca_data)
print(np.unique(pred,return_counts=True))
plt.scatter(pca_data[:,0],pca_data[:,1],c=pred)
plt.scatter(pca_data[:,0],pca_data[:,1],c=label, s=2.0)
plt.show()
```
