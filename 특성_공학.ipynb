```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_sample = np.array([[1,2]])
pf = PolynomialFeatures(degree=2,include_bias=False)
pf.fit(x_sample)
xt = pf.transform(x_sample)
xt
```

```python
x2 = np.array([[3,4]])
xt = pf.transform(x2)
xt
```

```python
x = np.array([[np.random.uniform(-10,10),
               np.random.uniform(-10,10)]
               for _ in range(100)])
plt.plot(x[:,0],x[:,1],'r.')
plt.show()
```

```python
x_1 = x[:,0]
x_2 = x[:,1]
y = 2*(x_1**2)+3*(x_1*x_2)+4*x_1+2.3-3.5*(x_2**2)+7 + np.random.uniform(-100,100,size=100)
```

```python
y
```

```python
plt.plot(x_1,y,'r.')
plt.show()
```

```python
xt = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(xt,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
plt.plot(y_test,'bo')
plt.plot(pred,'r.')
plt.show()
```

```python
x_train2,x_test2,y_train2,y_test2 = train_test_split(x,y)
model2 = LinearRegression()
model2.fit(x_train2,y_train2)
pred2 = model2.predict(x_test2)
```

```python
plt.plot(y_test2,'bo')
plt.plot(pred2,'r.')
plt.show()
```

```python
model.score(x_test,y_test)
```

```python
model2.score(x_test2,y_test2)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,pred)
```

```python
r2_score(y_test2,pred2)
```

```python
model.coef_, model.intercept_
```
